```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import imageio
```

```python
image = imageio.imread("C:/Users/oekpom/Desktop/Datasets/Machine Learning/Lesson 7/dogs.jpeg")
plt.figure(figsize = (15,8))
plt.imshow(image)
```

```python
image.shape
```

```python
image.ndim
```

```python
x, y, z = image.shape
image_2d = image.reshape(x*y, z)
image_2d.shape
```

```python
image_2d.ndim
```

```python
kmeans_cluster = KMeans(n_clusters=3)
kmeans_cluster.fit(image_2d)
cluster_centers = kmeans_cluster.cluster_centers_
cluster_labels = kmeans_cluster.labels_
```

```python
import numpy as np
plt.figure(figsize = (15,8))
data = cluster_centers[cluster_labels].reshape(x, y, z)
image_8bit = np.uint8(data)
plt.imshow(image_8bit)
```
